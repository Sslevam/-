```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Заданный столбец данных
column = 3  


filename = "12_Пермь.csv"
with open(filename, "r") as f:
    reader = csv.reader(f)
    data = [row for row in reader]

data = np.array(data)

# Определение диапазона строк для анализа
start = 21  # Исключение заголовка
end = start + 12

# Выбор значений
x = []
y = []
for i in range(start, end):
    if data[i, column] != "999.9":
        x.append(float(data[i, 0]))
        y.append(float(data[i, column]))

# Построение интерполяционный полином Лагранжа
p = np.poly1d(np.polyfit(x, y, len(x)-1))

# Полином Лагранжа 
h = 0.1  # Шаг сетки
x_grid = np.arange(min(x), max(x)+h, h)
plt.plot(x, y, "ro", label="Данные")
plt.plot(x_grid, p(x_grid), "b-", label="Полином Лагранжа")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()


# Выбор значений, отличных от 999.9
x = []
y = []
for i in range(start, end):
    if data[i, column] != "999.9":
        x.append(float(data[i, 0]))
        y.append(float(data[i, column]))

# Построение интерполяционного полинома первой формулой Ньютона
n = len(x)
c = np.zeros(n)
c[0] = y[0]
for j in range(1, n):
    for i in range(n-1, j-1, -1):
        y[i] = (y[i] - y[i-1]) / (x[i] - x[i-j])
    c[j] = y[j]

p = np.poly1d(c[::-1])

# график интерполяции Ньютона 
h = 0.1
x_grid = np.arange(min(x), max(x)+h, h)
plt.plot(x, y, "ro", label="Данные")
plt.plot(x_grid, p(x_grid), "b-", label="Интерполяция Ньютона")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()




# Выбор значений для интерполяции
x = []
y = []
for i in range(start, end):
    if data[i, column] != "999.9":
        x.append(float(data[i, 0]))
        y.append(float(data[i, column]))

# Разность конечных разностей
def diff_quotients(x, y):
    n = len(x)
    F = np.zeros((n, n))
    F[:,0] = y
    for j in range(1,n):
        for i in range(j,n):
            F[i,j] = (F[i,j-1]-F[i-1,j-1])/(x[i]-x[i-j])
    return F

# Вычисление значений полинома
def newton2(x, y, z):
    F = diff_quotients(x, y)
    n = len(x)
    p = F[n-1,n-1]
    for k in range(2, n+1):
        prod = 1
        for i in range(n-k+1, n):
            prod *= (z - x[i])
        p += F[n-1,k-1] * prod
    return p

# График 2 формулы Ньтона 
h = 0.1  
x_grid = np.arange(min(x), max(x)+h, h)
plt.plot(x, y, "ro", label="Данные")
plt.plot(x_grid, [newton2(x, y, z) for z in x_grid], "b-", label="Вторая интерполяционная формула Ньютона")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
